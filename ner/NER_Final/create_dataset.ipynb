```python
import pandas as pd
import re
import string
```

```python
def preprocessing(berita):
    s = str(berita)
    s = berita.lower()
    s = s.replace('\n', ' ')
    s = s.replace('\r', ' ')
    s = re.sub(r'[^a-zA-Z0-9\s]', ' ', s)
    s = s.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    s = s.strip()
    s = s.translate(str.maketrans("","", string.punctuation))
    tokens = [token for token in s.split(" ") if token != ""]
    T = [t for t in tokens]
    return T
```

```python
df_train = pd.read_csv('df_train.csv')
df_train.head()
```

```python
for x in range(0, df_train.shape[0]):
  text = preprocessing(df_train.iloc[x, -2])
  df_train.iloc[x, -2] = (' '.join(text))
```

```python
news = []
for x in range(0, df_train.shape[0]):
  news.append([df_train.iloc[x, -3], df_train.iloc[x, -2]])

print(len(news))
```

```python
document_results = []
for x in range(0, len(news)):
  document_results.append([
    x+1,
    news[x][0],
    news[x][1],
    '',
    '',
    '',
    '',
    ''
  ])

writer = pd.DataFrame(document_results, columns=['No', 'Tanggal', 'Berita', 'CD (Korban dll)', 'What (Topik)', 'Rendah', 'Sedang', 'Tinggi'])
writer.to_excel('dataset_keparahan.xlsx', index=False)
```
